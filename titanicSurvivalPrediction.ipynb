```python
data = data.drop('Cabin', axis=1)
```

```python
# Fill missing values in 'Age' with the mean age
data['Age'].fillna(data['Age'].mean(), inplace=True)

# Fill missing values in 'Embarked' with the most frequent value
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(data.isnull(), cbar=False, cmap='viridis')
plt.title('Null Values Heatmap')
plt.show()
```

```python
# Calculate the correlation matrix
corr_matrix = data[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].corr()

# Plot heatmap of correlation
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
categorical_variables = data.select_dtypes(include='object').columns.tolist()
```

```python
data_encoded = pd.get_dummies(data, columns=categorical_variables, drop_first=True)
```

```python
data_encoded.head()
```

```python
# Define features (X) and target variable (y)
X = data_encoded.drop('Survived', axis=1)  # All columns except 'price'
y = data_encoded['Survived']  # The 'price' column is the target variable
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression

# Initialize the Logistic Regression model
model = LogisticRegression(max_iter=1000)

# Train the model
model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
test_data = pd.read_csv("/content/test.csv")
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(test_data.isnull(), cbar=False, cmap='viridis')
plt.title('Null Values Heatmap')
plt.show()
```

```python
test_data = test_data.drop('Cabin', axis=1)
```

```python
# Fill missing values in 'Age' with the mean age
test_data['Age'].fillna(test_data['Age'].mean(), inplace=True)

# Fill missing values in 'Fare' with the mean fare
test_data['Fare'].fillna(test_data['Fare'].mean(), inplace=True)
```

```python
categorical_variables = test_data.select_dtypes(include='object').columns.tolist()
```

```python
test_data_encoded = pd.get_dummies(test_data, columns=categorical_variables, drop_first=True)
```

```python
test_data_encoded.head()
```

```python
# Get the feature names from the training data
training_features = model.feature_names_in_

# Reindex the test data to match the training data features
test_data_encoded = test_data_encoded.reindex(columns=training_features, fill_value=0)

# Now you can predict using the reindexed test data
y_pred = model.predict(test_data_encoded)
```

```python
# Create a DataFrame with PassengerId and predictions
submission_df = pd.DataFrame({'PassengerId': test_data_encoded['PassengerId'], 'Survived': y_pred})

# Save the DataFrame to a CSV file
submission_df.to_csv('submission.csv', index=False)
```
